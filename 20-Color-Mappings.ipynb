```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/computer vision/DATA/sunset.PNG')
```

```python
plt.imshow(img)
```

### Converting to Different Colorspaces

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

**Converting to HSV**
https://en.wikipedia.org/wiki/HSL_and_HSV

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img)
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/computer vision/DATA/sunset.PNG')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img)
```
