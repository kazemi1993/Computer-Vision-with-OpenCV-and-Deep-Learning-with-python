```python
import cv2
import matplotlib.pyplot as plt

# Correct file path with leading slash
img_bgr = cv2.imread('/content/drive/MyDrive/Colab Notebooks/computer vision/DATA/ai.jpg')

# Check if image loaded properly
if img_bgr is None:
    print("Image not found. Check the file path.")
else:
    # Convert BGR to RGB
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # Show image with matplotlib
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()
```

The image has been correctly loaded by openCV as a numpy array, but the color of each pixel has been sorted as BGR.
Matplotlib's plot expects an RGB image so, for a correct display of the image, it is necessary to swap those channels.
This operation can be done either by using openCV conversion functions cv2.cvtColor() or by working directly with the numpy array.

```python
import cv2
import matplotlib.pyplot as plt

# Correct path with leading slash and grayscale flag
img_gray = cv2.imread('/content/drive/MyDrive/Colab Notebooks/computer vision/DATA/ai.jpg', cv2.IMREAD_GRAYSCALE)

# Check if image loaded correctly
if img_gray is None:
    print("Image not found. Check the file path.")
else:
    # Display grayscale image with appropriate colormap
    plt.imshow(img_gray, cmap='gray')
    plt.axis('off')
    plt.show()
```

## Resize Images

```python
img_rgb.shape
# width, height, color channels
```

```python
img =cv2.resize(img_rgb,(1300,275))
```

```python
plt.imshow(img)
```

### By ratio

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img =cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

### Flipping Images

```python
# Along central x axis
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
# Along central y axis
new_img = cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
# Along both axis
new_img = cv2.flip(new_img,-1)
plt.imshow(new_img)
```
