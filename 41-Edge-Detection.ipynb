# Canny Edge Detection
https://en.wikipedia.org/wiki/Canny_edge_detector

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/dog-with-kid.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=200, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)
```

```python
plt.imshow(edges)
```

# blur

```python
blurred_image = cv2.blur(img,ksize=(5,5))
```

```python
plt.imshow(blurred_image)
```

```python
edges = cv2.Canny(image=blurred_image, threshold1=0, threshold2=255)
```

```python
plt.imshow(edges)
```
