# Gradients


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/periodictable.png',0)
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
display_img(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
display_img(sobely)
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
display_img(laplacian)
```

## Combining Previous Ideas

Let's play around with these images with some of the other ideas we've already seen!

### Blending Images

```python
blended = cv2.addWeighted(src1=sobelx,alpha=0.5,src2=sobely,beta=0.5,gamma=0)
```

```python
display_img(blended)
```

### Thresholds

```python
ret,th1 = cv2.threshold(blended,140,255,cv2.THRESH_BINARY_INV)
display_img(th1)
```

```python
ret,th1 = cv2.threshold(img,160,255,cv2.THRESH_BINARY)
display_img(th1)
```
